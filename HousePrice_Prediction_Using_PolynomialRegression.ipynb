```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myData=pd.read_csv('Housing_Price.csv')
myData
```

```python
x=myData.iloc[:,0:1].values
x
```

```python
y=myData.iloc[:,1].values
y
```

```python
plt.scatter(x,y,color='red')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
createmyModel=LinearRegression()
createmyModel.fit(x,y)
```

```python
plt.scatter(x,y,color='b')
plt.plot(x,createmyModel.predict(x),color='r')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
createpolyModel=PolynomialFeatures(degree=3)
x_poly=createpolyModel.fit_transform(x)
x_poly
```

```python
creatModelWithPoly=LinearRegression()
creatModelWithPoly.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='blue')
plt.scatter(x,creatModelWithPoly.predict(x_poly),color='r')
plt.show()
```

```python
creatModelWithPoly.score(x_poly,y)
```
